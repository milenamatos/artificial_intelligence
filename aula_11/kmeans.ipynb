```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale

%matplotlib inline

aggregationData = pd.read_csv("Aggregation.csv")

km = KMeans(n_clusters=5, random_state=0)
y_predicted = km.fit_predict(aggregationData[['labelA', 'labelB']])
aggregationData['cluster'] = y_predicted

for i in range(0,5):
    _df = aggregationData[aggregationData.cluster==i]
    plt.scatter(_df.labelA,_df.labelB, color=np.random.rand(3))
```

```python
#D31
d31Data = pd.read_csv("D31.csv")

km = KMeans(n_clusters=10, random_state=0)
y_predicted = km.fit_predict(d31Data[['labelA', 'labelB']])
d31Data['cluster'] = y_predicted

for i in range(0,10):
    _df = d31Data[d31Data.cluster==i]
    plt.scatter(_df.labelA,_df.labelB, color=np.random.rand(3))
```

```python
#PathBased
pathBasedData = pd.read_csv("PathBased.csv")

pathBasedData = pathBasedData.drop(columns=['labelC'])
X = scale(pathBasedData)

km = KMeans(n_clusters=3, random_state=0)
y_predicted = km.fit_predict(X)

pathBasedData['cluster'] = y_predicted

for i in range(0,3):
    _df = pathBasedData[pathBasedData.cluster==i]
    plt.scatter(_df.labelA,_df.labelB, color=np.random.rand(3))
```
